## 處理SPOT/tag表格

```python
def showResult(result):
    if len(result)>0:
        columns=list(result[0].keys())
        return pd.DataFrame(result,columns=columns)
    else:
        return None
```

## 匯入

```python
import pandas as pd
from sqlalchemy import create_engine
# read CSV file
df = pd.read_csv('Spot_table.csv', encoding='utf-8')
#csv 匯入 DB
try:
    engine = create_engine('mysql+pymysql://root:iii@mysql/test')
    with engine.connect() as conn, conn.begin():
        df.to_sql('spot', conn, if_exists='replace', index=True)


    #build connenction
    import pymysql
    connection = pymysql.connect(host='mysql', 
                                 user='root',
                                 password='iii',
                                 charset='utf8mb4',
                                 database='test',
                                 cursorclass=pymysql.cursors.DictCursor)

    with connection.cursor() as cursor:
        sql = "USE `test`"
        cursor.execute(sql)
        sql_query = "SELECT * FROM `spot`"
#         sql_query = "Drop Table `spot`"
        cursor.execute(sql_query)
        result = cursor.fetchall()
        showResult(result)

except:
    connection.rollback() #Rollback the changes on error

finally:
    connection.close()
    showResult(result)

result
```
