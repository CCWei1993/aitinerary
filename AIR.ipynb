```python
df_Dur = pd.concat([df['DDur'],df['satRDur'],df['friRDur'],df['thuRDur'],df['wedRDur'], df['tueRDur'], df['monRDur'],df['sunRDur']], axis=1)

for i in df_Dur:
    df[i] = pd.DatetimeIndex(df[i]).hour*60 + pd.DatetimeIndex(df[i]).minute
df_Dur
```

```python
df_Day = pd.concat([df['DEtd'], df['DEta'], df['satREtd'], df['satREta'], df['friREtd'],df['friREta'], df['thuREtd'], 
                   df['thuREta'], df['wedREtd'], df['wedREta'], df['tueREtd'], df['tueREta'], df['monREtd'], df['monREta'],
                   df['sunREtd'],df['sunREta']], axis=1)
```

```python
for i in df_Day:
    df[i + '_day'] = df[i].apply(lambda s : s.split(' ')[0])
    df[i + '_time'] = df[i].apply(lambda s : s.split(' ')[1])
```

```python
df = df.drop(columns=['DEtd', 'DEta', 'satREtd','satREta', 'friREtd', 'friREta', 'thuREtd', 'thuREta', 'wedREtd',
                     'wedREta', 'tueREtd','tueREta', 'monREtd', 'monREta', 'sunREtd','sunREta'])
```

```python
def mapping_df_types(df):
    dtypedict = {}
    for i, j in zip(df.columns, df.dtypes):
        if "object" in str(j):
            dtypedict.update({i: NVARCHAR(length=255)})
        if "float" in str(j):
            dtypedict.update({i: Float(precision=2, asdecimal=True)})
        if "int" in str(j):
            dtypedict.update({i: Integer()})
        if "timedelta64[ns]" in str(j):
            dtypedict.update({i: DATETIME()})
    return dtypedict
```

```python
from sqlalchemy.types import NVARCHAR, Float, Integer, DATETIME
from sqlalchemy import create_engine
engine = create_engine("mysql+pymysql://root:iii@mysql/test".format('root', 'iii', 'host:mysql', 'test'))
con = engine.connect()

dtypedict = mapping_df_types(df)
df.to_sql(name='air', con=con, if_exists='append', index=False, dtype=dtypedict)
```
